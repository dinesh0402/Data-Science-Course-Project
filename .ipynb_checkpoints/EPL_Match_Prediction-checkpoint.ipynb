## Data Preparation for Initial ML Model.

```python
# Convert date to datetime object using Pandas for future use.
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches.dtypes
```

Lets convert the "venue" attribute of the dataset to machine understandable code.

The "venue" attribute determines whether the team is playing in "Home" or "Away" from home (if you have seen football, this is known as "Home Ground Advantage").

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
```

```python
matches.head()
```

Also, lets use an "opp_code" or short for opponent code for categorization purposes in the ML model.

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches.head()
```

It may be a case that some teams play better than others at a particular time (hour in this case) of the day.

```python
matches["hour"] = matches["time"].str.replace(":.+","",regex=True).astype("int")
```

```python
matches.head()
```

Lets add a "day_code" column to the dataset to determine on which day of the week is the match played on (6-Sunday & 0-Monday).

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches.head()
```

Now the target of this prediction is the the outcome of the match i.e., Win(W), Loss(L) or Draw(D).

I assuming the target values to be 0 or 1:
- 0 means Loss/Draw.
- 1 means a Win.

```python
matches["target"] = (matches["result"] == "W").astype("int")
```

```python
matches.head()
```

***

## Building the initial ML model using Random Forest Classifier.

The reason I'm using Random Forest Classifier is that it picks up non-linearities in the data.

For example, the "opp_code" attribute isn't linearly related to any other attributes. Higher opp_code doesn't necessarily mean more difficult opponent.

This can't be picked up by linear models & hence RFC.

```python
# n_estimators -> #individual decision trees (part of the Random Forest) we want to train
# min_samples_split -> #samples we want to have in a leaf of a decision tree (higher value => lesser overfitting but lower accuracy)-> experiment this value.

rf = RandomForestClassifier(n_estimators=50, min_samples_split = 10, random_state = 1)
```

```python
train = matches[matches["date"] < '2022-01-01']
```

```python
test = matches[matches["date"] > '2022-01-01']
```

### <u>Reasoning behind the train & test split:</u>
If we can evaluate the performance of the algorithm (error rate, etc.) on the same dataset, our algorithm is gonna work awesome. 

```python
predictors = ["venue_code","opp_code","hour","day_code"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
acc = accuracy_score(test["target"], preds)
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual = test["target"], prediction = preds))
```

```python
pd.crosstab(index = combined["actual"], columns = combined["prediction"])
```

We see that when the model predicted a loss, it was more often right (141 times) than wrong (76 times).

But when the model predicted a win, it was more often wrong (31 times) than it was right (28 times).

Lets see how many percentage times the team actually won when it was predicted a win.

```python
precision_score(test["target"], preds)
```

## Improving Precision using rolling averages.

Lets add more predictors to improve our algorithm performance.

for this, we split up the dataset into team-wise group to compute rolling averages for team performance.

```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester City")
```

```python
group
```

```python
#  Function to calculate Rolling Averages.

def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    
    # Take the current week out for the the prediction (closed = 'left')
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    
    group[new_cols] = rolling_stats
    
    # Drop any missing values in the data, like if were are in Week 2, we can't get the info about the previous 3 weeks (as only 2 weeks have been played).
    group = group.dropna(subset = new_cols)
    return group
```

```python
cols = ["gf","ga","sh","sot","dist","fk","pk","pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

Lets look at a sample of rolling averages for the Manchester City.

```python
rolling_averages(group, cols, new_cols)
```

```python
# Rolling averages computed team-wise.
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
# Give the increasing order of indices.
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

## Retraining the ML Model`

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual = test["target"], predicted = preds), index = test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined , precision = make_predictions(matches_rolling, predictors+new_cols)
```

```python
precision
```

We have improved the precision of the model by quite a lot, from 47% to 62.5%.

```python
combined
```

```python
combined = combined.merge(matches_rolling[["date","team","opponent","result"]], left_index=True, right_index=True)
```

## Combining Home and Away predictions.

Lets see how the algorithm performed in predicting both sides of the match. We might have predicted a win for both sides of the match which isn't quite right (both the teams can't win in the same match). Let's resolve such issues.

Lets normalize the team names for consistency (like Wolverhampton Wanderers to Wolves).

```python
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {
    "Brighton and Hove Albion": "Brighton",
    "Manchester United": "Manchester Utd",
    "Newcastle United": "Newcastle Utd",
    "Tottenham Hotspur": "Tottenham",
    "West Ham United": "West Ham",
    "Wolverhampton Wanderers": "Wolves"
}

mapping = MissingDict(**map_values)
```

```python
mapping["Manchester United"]
```

```python
combined["new_team"] = combined["team"].map(mapping)
```

```python
combined
```

```python
merged = combined.merge(combined, left_on=["date","new_team"], right_on=["date","opponent"])
```

```python
merged
```

```python
# The no. of times a team was predicted to win & other team predicted to lose.
merged[(merged["predicted_x"] == 1) & (merged["predicted_y"] == 0)]["actual_x"].value_counts()
```

### Thus we see that the accuracy of the model improved from 61.23% to 67.5% just by involving both sides of the match & not just one.
